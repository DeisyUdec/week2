```python
#Desarrollado por Deisy Murcia
import numpy as np
def step(z):
  return 1 if z>=0 else 0
def neurona (x1, x2, w1, w2, b):
  z= x1*w1+x2*w2+b
  y=step(z)
  return z,y

#valores base
w1, w2, b= -0.5,1,-1.5

casos=[(0,0),(0,1),(1,0),(1,1)]
for x1, x2 in casos:
  z,y =neurona(x1,x2,w1,w2,b)
  print (f"Entrada=({x1,},{x2} z={z:.1f} salida {y})")
```
