```python
#importa las librerias utilizar elemetos de la IA y para crear un arreglo
import tensorflow as tf
import numpy as np
```

```python
#Datos que se utiliza para aprender
# se crear una entrada con valores de %
porcentaje=np.array([12,25,8,15],dtype=float)
#se crea una lista de los elementos de salida
radianes=np.array([20,41.6,13.3,25],dtype=float)
```

```python
#se crean las capas densas de kera
capa=tf.keras.layers.Dense(units=1,input_shape=[1])
modelo=tf.keras.Sequential([capa])
```

```python
#se crea el modelo de aprendizaje automatico ajustando los peso y los sesgos del ejerccio
modelo.compile(
    #ingesa la tasa de aprendizaje
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
```

```python
print("comenzamos entrenamiento..........")
#ingresa los datos de entrada y los de saalida y cuantas vueltas quiero que
#lo intente el verbose=false es para imprimir solo las opciones verdaderas
historial=modelo.fit(porcentaje,radianes,epochs=1000, verbose=False)
print("modelo entrenado")
```

```python
#crear la funcion de perdida
import matplotlib.pyplot as plt
plt.xlabel("# epoca")
plt.ylabel("magnitud de perdida")
plt.plot(historial.history["loss"])
```

```python
print("hagamos una predicción")
#es imporatnte colocar el valor del prcentaje y el tipo de dato
resultado=modelo.predict(np.array([60],dtype=float))
print("el resultado es"+str(resultado)+" radianes")
```
